# Transfer Learning with Pretrained CNNs for Image Classification  
### Using Fashion-MNIST Dataset  

---

## Objective
This tutorial demonstrates how transfer learning improves image classification performance by comparing:
- A baseline CNN trained from scratch
- A pretrained CNN using feature extraction
- A fine-tuned pretrained CNN
- Data augmentation impact

### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import time

from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

np.random.seed(42)
tf.random.set_seed(42)
```

### Load Dataset

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)

class_names = ['T-shirt', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle Boot']
```

### Sample Images

```python
plt.figure(figsize=(8,8))

for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(X_train[i].squeeze(), cmap='gray')
    plt.title(class_names[y_train[i]])
    plt.axis('off')

plt.suptitle("Sample Fashion-MNIST Images", fontsize=14)
plt.tight_layout()
plt.savefig('sample_images.png')
plt.show()
```

### Baseline CNN

```python
def build_baseline():
    model = models.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
        layers.MaxPooling2D(),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

baseline_model = build_baseline()

history_baseline = baseline_model.fit(
    X_train, y_train,
    epochs=5,
    validation_data=(X_test, y_test),
    verbose=1
)
```

### Prepare for Transfer Learning

```python
X_train_rgb = np.repeat(X_train, 3, axis=-1)
X_test_rgb = np.repeat(X_test, 3, axis=-1)

X_train_rgb = tf.image.resize(X_train_rgb, (96,96))
X_test_rgb = tf.image.resize(X_test_rgb, (96,96))
```

### Transfer Learning Model

```python
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(96,96,3))
base_model.trainable = False

transfer_model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

transfer_model.compile(optimizer='adam',
                       loss='sparse_categorical_crossentropy',
                       metrics=['accuracy'])

history_transfer = transfer_model.fit(
    X_train_rgb, y_train,
    epochs=5,
    validation_data=(X_test_rgb, y_test),
    verbose=1
)
```

### Fine Tuning

```python
base_model.trainable = True

for layer in base_model.layers[:-20]:
    layer.trainable = False

transfer_model.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
                       loss='sparse_categorical_crossentropy',
                       metrics=['accuracy'])

history_finetune = transfer_model.fit(
    X_train_rgb, y_train,
    epochs=5,
    validation_data=(X_test_rgb, y_test),
    verbose=1
)
```

### Accuracy Comparision

```python
plt.figure(figsize=(8,5))

plt.plot(history_baseline.history['val_accuracy'], label='Baseline', color='blue')
plt.plot(history_transfer.history['val_accuracy'], label='Transfer', color='orange')
plt.plot(history_finetune.history['val_accuracy'], label='Fine-Tuned', color='green')

plt.title("Validation Accuracy Comparison")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.savefig('accuracy_comparison.png')
plt.show()
```

### Loss Comparision

```python
plt.figure(figsize=(8,5))

plt.plot(history_baseline.history['val_loss'], label='Baseline', color='blue')
plt.plot(history_transfer.history['val_loss'], label='Transfer', color='orange')
plt.plot(history_finetune.history['val_loss'], label='Fine-Tuned', color='green')

plt.title("Validation Loss Comparison")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.savefig('loss_comparison.png')
plt.show()
```

### Bar Chart Comparision

```python
baseline_acc = history_baseline.history['val_accuracy'][-1]
transfer_acc = history_transfer.history['val_accuracy'][-1]
finetune_acc = history_finetune.history['val_accuracy'][-1]

plt.figure(figsize=(6,5))
plt.bar(['Baseline','Transfer','Fine-Tuned'],
        [baseline_acc, transfer_acc, finetune_acc])

plt.title("Final Accuracy Comparison")
plt.ylabel("Accuracy")
plt.ylim(0,1)
plt.savefig('bar_chart.png')
plt.show()
```

### Confusion Matrix

```python
y_pred = np.argmax(transfer_model.predict(X_test_rgb), axis=1)

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm)

disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.savefig('confusion_matrix.png')
plt.show()
```

### Data Augmentation

```python
data_aug = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1)
])

aug_model = models.Sequential([
    data_aug,
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

aug_model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

history_aug = aug_model.fit(
    X_train_rgb, y_train,
    epochs=5,
    validation_data=(X_test_rgb, y_test),
    verbose=1
)
```

### Augumentaion Plot

```python
plt.figure(figsize=(8,5))

plt.plot(history_transfer.history['val_accuracy'], label='Transfer', color='orange')
plt.plot(history_aug.history['val_accuracy'], label='Augmented', color='purple')

plt.title("Effect of Data Augmentation")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.savefig('augmentation_plot.png')
plt.show()
```

### Predictions

```python
preds = np.argmax(transfer_model.predict(X_test_rgb), axis=1)

plt.figure(figsize=(10,6))

for i in range(8):
    plt.subplot(2,4,i+1)
    plt.imshow(X_test[i].squeeze(), cmap='gray')

    true = class_names[y_test[i]]
    pred = class_names[preds[i]]

    color = 'green' if true == pred else 'red'
    plt.title(f"T:{true}\nP:{pred}", color=color)
    plt.axis('off')

plt.suptitle("Prediction Analysis")
plt.tight_layout()
plt.savefig('prediction_analysis.png')
plt.show()
```

## Conclusion
- Transfer learning significantly improves accuracy compared to baseline CNN.
- Fine-tuning further enhances performance by adapting pretrained features.
- Data augmentation improves generalisation.
- Transfer learning reduces training time while achieving higher accuracy.
